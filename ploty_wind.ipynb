Paso 1: Carga de Datos Climáticos
Plotly incluye datasets de prueba. Utilizaremos wind(), que contiene información sobre la dirección del viento y su intensidad.



```python
import plotly.express as px

# Cargar el dataset integrado
wind_df = px.data.wind()

# Inspección inicial
print(wind_df.head())
print(wind_df.info())
```

Paso 2: Gráfico de Dispersión Polar (Scatter Polar)
Este gráfico es ideal para ver la distribución de puntos individuales. Usaremos r para la magnitud y theta para la dirección angular.

```python
fig1 = px.scatter_polar(wind_df, 
                        r="strength", 
                        theta="direction", 
                        color="strength",
                        symbol="direction",
                        size="frequency",
                        title="Dispersión Polar: Fuerza vs Dirección del Viento")
fig1.show()
```

Paso 3: Gráfico de Líneas Polar (Line Polar)
Útil para observar tendencias continuas o cierres de ciclos. En este caso, conectaremos los niveles de fuerza por dirección.

```python
fig2 = px.line_polar(wind_df, 
                      r="strength", 
                      theta="direction", 
                      color="strength",
                      line_close=True, # Cierra el círculo de la línea
                      title="Tendencia Polar: Intensidad por Dirección")
fig2.show()
```

Paso 4: Gráfico de Barras Polar (Wind Rose)
Este es el estándar para representar la dirección del viento (Rosa de los vientos). Las barras se extienden desde el centro según su magnitud.

```python
fig3 = px.bar_polar(wind_df, 
                    r="frequency", 
                    theta="direction", 
                    color="strength",
                    color_discrete_sequence=px.colors.sequential.Plasma_r,
                    title="Rosa de los Vientos: Frecuencia por Dirección")
fig3.show()
```

Paso 5: Personalización Avanzada y Exportación
Ajustaremos la apariencia del gráfico para un reporte profesional y guardaremos los resultados como archivos HTML interactivos.

```python
# Ejemplo de personalización del diseño
fig3.update_layout(
    font_size=12,
    legend_font_size=10,
    polar_angularaxis_rotation=90, # Rota el gráfico 90 grados
)

# Guardar como HTML
fig1.write_html("polar_scatter.html")
fig2.write_html("polar_line.html")
fig3.write_html("polar_bar.html")
```

1. ¿Qué patrones de dirección del viento predominantes observas en el gráfico de barras polar?
En un gráfico de barras polar (comúnmente llamado "rosa de los vientos"), el patrón predominante se identifica observando cuál de las "barras" o "pétalos" es el más largo. La longitud total de la barra apuntando hacia una dirección cardinal específica (como Norte, Este, Suroeste) indica que el viento sopló desde esa dirección con mayor frecuencia. Además, gracias a la división por colores en esa misma barra, puedes observar no solo que el viento fue frecuente desde ahí, sino también la proporción de las diferentes intensidades de viento asociadas a esa dirección.

2. ¿Cuál de las tres representaciones (dispersión, línea o barra) consideras que comunica mejor la intensidad del viento y por qué?
El gráfico de barras polar (px.bar_polar) es, con diferencia, el que mejor comunica la información del viento. ¿Por qué? Porque utiliza áreas rellenas y apiladas. Cada segmento de color dentro de una barra representa un rango de intensidad (fuerza del viento), y la barra al estar apilada suma visualmente la frecuencia total de manera clara y directa.

El gráfico de dispersión (scatter) no es ideal porque los puntos pueden superponerse (overplotting), lo que dificulta ver el volumen real acumulado de los datos.
El gráfico de líneas (line) resulta confuso para este propósito, ya que conectar frecuencias de diferentes intensidades dibuja telarañas o cruces en el centro que son muy difíciles de interpretar intuitivamente.
3. ¿Por qué es necesario usar un sistema de coordenadas polares en lugar de un gráfico de barras tradicional (rectangular) para datos de viento?
Es necesario porque la dirección del viento es una variable cíclica, direccional o periódica (va de 0° a 360°, donde los 360° conectan de nuevo con los 0°). Si usaras un gráfico de barras rectangular (cartesiano XY), tendrías un inicio y un fin (por ejemplo, el eje X empezando en el Norte y terminando en el Nor-Noroeste en el otro extremo). Esto rompe visualmente la continuidad espacial, alejando direcciones que en la vida real son físicamente adyacentes. El sistema de coordenadas polares mapea los datos en un círculo continuo de 360°, lo que imita la realidad física y la forma en que leemos una brújula.

4. Investiga: ¿Qué representan los argumentos r y theta en la función px.scatter_polar?
En px.scatter_polar (y en todas las funciones polares de Plotly), estos parámetros definen las dos coordenadas clave del sistema polar:

r (Radio / Radius): Representa la distancia desde el centro del gráfico hacia el exterior. En los datos de viento, r se asigna a la magnitud de los datos (como la frecuencia o la velocidad). Cuanto mayor es el valor numérico, más alejado del centro se graficará el dato. Es el equivalente aproximado al "eje Y" en un gráfico cartesiano.
theta (Ángulo / Theta): Representa el ángulo o la posición rotacional alrededor del círculo. Se mapea a la dirección desde la que sopla el viento (valores categóricos como 'N', 'NE', 'S' o valores numéricos en grados de 0 a 360). Es el equivalente aproximado al "eje X", pero envuelto en una forma circular.
